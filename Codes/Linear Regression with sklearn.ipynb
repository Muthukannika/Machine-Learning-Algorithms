```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("House Price Prediction.csv")
```

```python
print(df)
```

```python
plt.scatter(df.area,df.price,color="red",marker="*")
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
plt.scatter(df.area,df.price,color="red",marker="*")
plt.plot(df.area,model.predict(df[['area']]))
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
model.predict([[3300]])
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```
